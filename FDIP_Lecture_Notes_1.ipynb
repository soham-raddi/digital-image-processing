```python
# First import the required Python Libraries
import numpy as np
import matplotlib.pyplot as plt
from skimage import img_as_uint
from skimage.io import imshow, imread
from skimage.color import rgb2hsv
from skimage.color import rgb2gray
```

```python
array_l = np.array([[255, 0], [0, 255]])
imshow(array_l, cmap = 'gray');
```

```python
array_2 = np.array([[255, 0, 255],
[0, 255, 0],
[255, 0, 255]])
imshow(array_2, cmap = 'gray');
```

```python
array_spectrum = np.array([np.arange(0,255, 17),
np.arange(255,0,-17),
np.arange(0,255, 17)])
fig ,ax = plt.subplots(1, 2, figsize=(12,4))
ax[0].imshow(array_spectrum, cmap='gray')
ax[0].set_title('Arrange Generation')
ax[1].imshow(array_spectrum.T,cmap = 'gray')
ax[1].set_title('Transpose Generation')
```

```python
array_colors = np.array([[255, 0, 0],
[0, 255, 0],
[0, 0, 255]])
imshow(array_colors);
```

```python
jungfrau = imread('/jungfrau.jpg')
imshow(jungfrau);
```

```python
jungfrau.shape
```

```python
split_width = jungfrau.shape[1] // 3

fig, ax = plt.subplots(1, 3,
                       figsize = (15, 5),
                       sharey= True)
ax[0].imshow(jungfrau[:, 0:split_width])
ax[0].set_title('First Split')

ax[1].imshow(jungfrau[:, split_width:2*split_width])
ax[1].set_title('Second Split')

ax[2].imshow(jungfrau[:, 2*split_width:3*split_width])
ax[2].set_title('Third Split');
```

```python
imshow(jungfrau[95:200, 900:1000]);
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12,4), sharey=True)
ax[0].imshow(jungfrau[:,:,0], cmap='Reds')
ax[0].set_title('Red')
ax[1].imshow(jungfrau[:,:,1], cmap='Greens')
ax[1].set_title('Green')
ax[2].imshow(jungfrau[:,:,2], cmap='Blues')
ax[2].set_title('Blue');
```

```python
jungfrau_hsv = rgb2hsv(jungfrau)
fig, ax = plt.subplots(1, 3, figsize=(12,4), sharey= True)
ax[0].imshow(jungfrau_hsv[:,:,0], cmap='hsv')
ax[0].set_title('Hue')
ax[1].imshow(jungfrau_hsv[:,:,1], cmap='gray')
ax[1].set_title('Saturation')
ax[2].imshow(jungfrau_hsv[:,:,2],cmap = 'gray')
ax[2].set_title('values')
```

```python
jungfrau_gray = rgb2gray(jungfrau)
fig, ax = plt.subplots(1, 4, figsize=(16,4), sharey= True)
ax[0].imshow(jungfrau_gray, cmap='gray')
ax[0].set_title('Grayscale orignal')
ax[1].imshow(img_as_uint(jungfrau_gray>0.25), cmap='gray')
ax[1].set_title('greater than 0.25');
ax[2].imshow(img_as_uint(jungfrau_gray>0.50), cmap='gray')
ax[2].set_title('greater than 0.50');
ax[3].imshow(img_as_uint(jungfrau_gray>np.mean(jungfrau_gray)), cmap='gray')
ax[3].set_title('greater than mean');
```
