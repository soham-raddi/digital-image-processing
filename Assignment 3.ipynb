```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open('./Images/mountains.jpg') 
original_array = np.array(img)

# --- Display the original image ---
plt.figure(figsize=(6, 6))
plt.imshow(original_array)
plt.title(f"Original Image\nShape: {original_array.shape}")
plt.axis('off')
plt.show()
```

```python
sampling_factors = [2, 8, 32]

# Create a plot to show the images side-by-side
fig, axes = plt.subplots(1, len(sampling_factors) + 1, figsize=(16, 4))

# Show the original image on the far left
axes[0].imshow(original_array)
axes[0].set_title("Original")
axes[0].axis('off')

# Loop through the factors, downsample, and display each result
for i, factor in enumerate(sampling_factors):
    sampled_array = original_array[::factor, ::factor]
    axes[i+1].imshow(sampled_array)
    axes[i+1].set_title(f"Downsampled 1/{factor}")
    axes[i+1].axis('off')

plt.tight_layout()
plt.show()
```

```python
quantization_levels = [16, 8, 4, 2]

# Create a plot to show the images side-by-side
fig, axes = plt.subplots(1, len(quantization_levels) + 1, figsize=(20, 5))

# Show the original image
axes[0].imshow(original_array)
axes[0].set_title("Original (256 Levels)")
axes[0].axis('off')

for i, levels in enumerate(quantization_levels):
    step = 256 / levels
    quantized_array = (original_array // step) * step
    
    axes[i+1].imshow(quantized_array.astype(np.uint8))
    axes[i+1].set_title(f"Quantized ({levels} Levels)")
    axes[i+1].axis('off')

plt.tight_layout()
plt.show()
```
